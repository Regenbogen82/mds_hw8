Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

**Нормальное распределение**

```python
mu_norm_1 = 11.5 # средний возраст школьников
mu_norm_2 = 5.5 # средний чек в супермаркете в тысячах
mu_norm_3 = 10 # средний вес собак
mu_norm_4 = 23 # средний возраст призыва в армию
mu_norm_5 = 16.6 # средняя стоимость смартфона в России в 2019 году в тысячах
sigma_norm_1 = 3.45
sigma_norm_2 = 2.1
sigma_norm_3 = 5
sigma_norm_4 = 1
sigma_norm_5 = 6.2
```

```python
?sts.norm
```

```python
student_age = sts.norm(loc=mu_norm_1, scale=sigma_norm_1)
check_value = sts.norm(loc=mu_norm_2, scale=sigma_norm_2)
dogs_weight = sts.norm(loc=mu_norm_3, scale=sigma_norm_3)
army_age = sts.norm(loc=mu_norm_4, scale=sigma_norm_4)
smartphone_price = sts.norm(loc=mu_norm_5, scale=sigma_norm_5)
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(0, 30, 1000)
plt.plot(x, check_value.cdf(x))
plt.plot(x, student_age.cdf(x))
plt.plot(x, dogs_weight.cdf(x))
plt.plot(x, army_age.cdf(x))
plt.plot(x, smartphone_price.cdf(x))
plt.legend(['check_value', 'student_age', 'dogs_weight', 'army_age', 'smartphone_price'])
plt.title('Функция распределения')
plt.show()
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(0, 30, 1000)
plt.plot(x, check_value.pdf(x))
plt.plot(x, student_age.pdf(x))
plt.plot(x, dogs_weight.pdf(x))
plt.plot(x, army_age.pdf(x))
plt.plot(x, smartphone_price.pdf(x))
plt.legend(['check_value', 'student_age', 'dogs_weight', 'army_age', 'smartphone_price'])
plt.title('функция плотности распределения')
plt.show()
```

**Экспоненциальное распределение**

```python
?sts.expon
```

```python
lambda_1 = 1/365 # гарантийный срок - 1 год, то есть в среднем без отказа пылесос работает в течение одного года
lambda_2 = 1/2 # в среднем специалист колл-центра разговаривает с клиентом 2 минуты
lambda_3 = 1/10 # в среднем ожидание автобуса на остановке составляет 10 минут
lambda_4 = 1/1 # между двумя звонками в колл-центр проходит 1 минута
lambda_5 = 1/0.5 # каждые 30 секунд подается новая заявка на сделку на бирже в рабочее время
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(0, 10, 1000)
plt.plot(x, sts.expon(scale=lambda_2).cdf(x))
plt.plot(x, sts.expon(scale=lambda_4).cdf(x))
plt.plot(x, sts.expon(scale=lambda_5).cdf(x))
plt.legend(['разговор в колл-центре', 'интенсивность звонков колл-центра', 'заявки на бирже'])
plt.title('Функция распределения')
plt.show()
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(0, 5, 1000)
plt.plot(x, sts.expon(scale=lambda_2).pdf(x))
plt.plot(x, sts.expon(scale=lambda_4).pdf(x))
plt.plot(x, sts.expon(scale=lambda_5).pdf(x))
plt.legend(['разговор в колл-центре', 'интенсивность звонков колл-центра', 'заявки на бирже'])
plt.title('Функция плотности распределения')
plt.show()
```

**Распределение Стьюдента**

```python
?sts.t
```

```python
rv1 = sts.t(10, loc=36.9, scale=0.6) # температура жителей дома (выборка - 10 человек)
rv2 = sts.t(5, loc=6, scale=2) # средний возраст детей, ежедневно играющих на детской площадке (выборка - 5 детей)
rv3 = sts.t(15, loc=13, scale=3) # кол-во бойцов в онлайн игре (15 игр)
rv4 = sts.t(18, loc=20, scale=5) # кол-во колорадского жука на одном кусте картошки
rv5 = sts.t(11, loc=2, scale=0.5) # кол-во бракованных изделий в партии наушников
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(-6, 40, 1000)
plt.plot(x, rv1.cdf(x))
plt.plot(x, rv2.cdf(x))
plt.plot(x, rv3.cdf(x))
plt.plot(x, rv4.cdf(x))
plt.plot(x, rv5.cdf(x))
plt.legend(['температура жителей дома', 'средний возраст детей на площадке', 'кол-во бойцов в онлайн игре',
           'кол-во колорадского жука на одном кусте', 'кол-во бракованных изделий в партии наушников'])
plt.title('Функция распределения')
plt.show()
```

```python
plt.figure(figsize=[18, 5])
x = np.linspace(-6, 40, 1000)
plt.plot(x, rv1.pdf(x))
plt.plot(x, rv2.pdf(x))
plt.plot(x, rv3.pdf(x))
plt.plot(x, rv4.pdf(x))
plt.plot(x, rv5.pdf(x))
plt.legend(['температура жителей дома', 'средний возраст детей на площадке', 'кол-во бойцов в онлайн игре',
           'кол-во колорадского жука на одном кусте', 'кол-во бракованных изделий в партии наушников'])
plt.title('Функция плотности распределения')
plt.show()
```

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$, т.е. $P(3.5&lt;\xi&lt;7)$.

```python
С = 1 / (5 - 0)
uniform_rv = sts.uniform(0, 5)
```

```python
plt.figure(figsize=[15, 5])
x = np.linspace(-2, 8, 1000)
plt.plot(x, uniform_rv.pdf(x))
plt.title('Функция плотности распределения Хи')
plt.show()
```

```python
print('математическое ожидание переменной =', uniform_rv.mean())
```

```python
print('вероятность попадания переменной в отрезок от 3.5 до 7 =', '{:.2f}'.format((1 / 5) * (5 - 3.5)))
```

**Задание 3**

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
group_full = 54 + 4 + 28 
p_red = 54 / group_full
p_yellow = 4 / group_full
p_green = 28 / group_full
```

```python
import math
b_red = math.log2(1/p_red)
b_yellow = math.log2(1/p_yellow)
b_green = math.log2(1/p_green)
```

```python
print('Количество информации о том, что красный =', '{:.2f}'.format(b_red))
print('Количество информации о том, что желтый =', '{:.2f}'.format(b_yellow))
print('Количество информации о том, что зеленый =', '{:.2f}'.format(b_green))
print('Больше всего информации несет тот цвет, что горит меньше всего. В нашем случае желтый')
```
